```python
import matplotlib.pyplot as plt
```

```python
# 1. Average duration by attack type
df.groupby("attack_type")["duration"].mean().sort_values(ascending=False).plot(kind="bar", figsize=(6,4), color="gold")
plt.title("Average Duration by Attack Type")
plt.ylabel("Duration")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("avg_duration_by_attack.png")
plt.show()

# 2. Count of attack type by protocol (pie chart)
df.groupby("protocol")["attack_type"].count().plot(kind="pie", figsize=(6,4), autopct="%1.1f%%")
plt.title("Count of Attack Type by Protocol")
plt.ylabel("")
plt.tight_layout()
plt.savefig("attack_count_by_protocol.png")
plt.show()

# 3. Average source & destination bytes by attack type
avg_bytes = df.groupby("attack_type")[["src_bytes", "dst_bytes"]].mean()
avg_bytes.plot(kind="bar", figsize=(6,4), color=["gold", "blue"])
plt.title("Average Source & Destination Bytes by Attack Type")
plt.ylabel("Bytes")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("avg_bytes_by_attack.png")
plt.show()

# 4. Average packet count by attack type
df.groupby("attack_type")["packet_count"].mean().sort_values(ascending=False).plot(kind="bar", figsize=(6,4), color="gold")
plt.title("Average Packet Count by Attack Type")
plt.ylabel("Packet Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("avg_packet_by_attack.png")
plt.show()

# 5. Average failed logins by attack type
df.groupby("attack_type")["failed_logins"].mean().sort_values(ascending=False).plot(kind="bar", figsize=(6,4), color="gold")
plt.title("Average Failed Logins by Attack Type")
plt.ylabel("Failed Logins")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("avg_failed_logins_by_attack.png")
plt.show()
```
